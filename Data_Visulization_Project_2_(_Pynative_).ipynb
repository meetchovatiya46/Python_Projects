```python
# Exercise 3: Read all product sales data and show it  using a multiline plot


import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("company_sales_data.csv")

month_number=df["month_number"].tolist()

facecream_product=df["facecream"].tolist()
facewash_product=df["facewash"].tolist()
toothpaste_product=df["toothpaste"].tolist()
bathingsoap_product=df["bathingsoap"].tolist()
shampoo_product=df["shampoo"].tolist()
moisturizer_product=df["moisturizer"].tolist()


plt.plot(month_number,facecream_product, label="facecream_product", marker="o")
plt.plot(month_number,facewash_product, label="facewash_product", marker="o")
plt.plot(month_number,toothpaste_product, label="toothpaste_product", marker="o")
plt.plot(month_number,bathingsoap_product, label="bathingsoap_product", marker="o")
plt.plot(month_number,shampoo_product, label="shampoo_product", marker="o")
plt.plot(month_number,moisturizer_product, label="moisturizer_product", marker="o")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Sales Data")
plt.legend()

plt.savefig("sales_data.jpg", dpi=300,bbox_inches="tight")

plt.savefig("sales_data.pdf", dpi=300,bbox_inches="tight")

plt.show()
```

```python
# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("company_sales_data.csv")

toothpaste_product=df["toothpaste"].tolist()
month_number=df["month_number"].tolist()

plt.scatter(month_number,toothpaste_product, marker="o", label="Toothpaste sales data")
plt.legend()
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Toothpaste sales data per month")
plt.grid(linestyle="--", alpha=0.6)

plt.tight_layout()

plt.savefig("toothpaste_sales_data.jpg", dpi=300,bbox_inches="tight")

plt.savefig("toothpaste_sales_data.pdf", dpi=300,bbox_inches="tight")

plt.show()
```

```python
# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("company_sales_data.csv")

month_number = df["month_number"].tolist()
facecream_product = df["facecream"].tolist()
facewash_product = df["facewash"].tolist()


bar_width = 0.35

x = np.arange(len(month_number))

plt.bar(x - bar_width/2, facecream_product, width=bar_width, color='skyblue', label="FaceCream Sales")
plt.bar(x + bar_width/2, facewash_product, width=bar_width, color='orange', label="FaceWash Sales")

plt.xlabel("Month Number")
plt.ylabel("Sales Units (in thousands)")
plt.title("Face Cream vs Face Wash Monthly Sales Data")
plt.xticks(x, month_number)
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.savefig("face_cream_vs_face_wash.jpg", dpi=300,bbox_inches="tight")

plt.savefig("face_cream_vs_face_wash.pdf", dpi=300,bbox_inches="tight")

plt.show()
```

```python
# Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart


import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv("company_sales_data.csv")


totalsale = [
    df["facecream"].sum(),
    df["facewash"].sum(),
    df["toothpaste"].sum(),
    df["bathingsoap"].sum(),
    df["shampoo"].sum(),
    df["moisturizer"].sum()
]

labels = ["Facecream", "Facewash", "Toothpaste", "Bathingsoap", "Shampoo", "Moisturizer"]

colors = plt.cm.Set3.colors

plt.figure(figsize=(8, 6))
plt.pie(
    totalsale,
    labels=labels,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    shadow=True
)

plt.title("Total Sales Data for Last Year", fontsize=14, fontweight="bold")

plt.legend(
    labels,
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    title="Products",
    fontsize=10
)

plt.tight_layout()

plt.savefig("total_sales_data_pie.jpg", dpi=300,bbox_inches="tight")

plt.savefig("total_sales_data_pie.pdf", dpi=300,bbox_inches="tight")

plt.show()
```

```python
# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv("company_sales_data.csv")

bathingsoap_product=df["bathingsoap"].tolist()
facewash_product=df["facewash"].tolist()
month_number=df["month_number"].tolist()



# it shows 2 different raws
fig,axes=plt.subplots(2, figsize=(8,6))

# it shows 1 raws and 2 columns
# fig,axes=plt.subplots(1,2)

axes[0].plot(month_number,bathingsoap_product, color="green", marker="o")
axes[0].set_title("Bathingsoap Sales data")


axes[1].plot(month_number,facewash_product, color="red", marker="o")
axes[1].set_title("Facewash Sales data")

plt.tight_layout()

plt.savefig("subplot_ex9.jpg", dpi=300,bbox_inches="tight")

plt.savefig("subplot_ex9.pdf", dpi=300,bbox_inches="tight")

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("company_sales_data.csv")

month_number = df["month_number"].tolist()

facecream_product = df["facecream"].tolist()
facewash_product = df["facewash"].tolist()
toothpaste_product = df["toothpaste"].tolist()
bathingsoap_product = df["bathingsoap"].tolist()
shampoo_product = df["shampoo"].tolist()
moisturizer_product = df["moisturizer"].tolist()

plt.figure(figsize=(10, 6))
plt.stackplot(
    month_number,
    facecream_product,
    facewash_product,
    toothpaste_product,
    bathingsoap_product,
    shampoo_product,
    moisturizer_product,
    labels=["Facecream", "Facewash", "Toothpaste", "Bathingsoap", "Shampoo", "Moisturizer"],
    colors=plt.cm.Set3.colors
)

plt.title("All Product Sales Data Using Stack Plot", fontsize=14, fontweight="bold")
plt.xlabel("Month Number")
plt.ylabel("Sales Units")

plt.legend(loc="upper left")

plt.grid(alpha=0.3)
plt.tight_layout()

plt.savefig("all_product_sales_data_stackplot.jpg", dpi=300,bbox_inches="tight")

plt.savefig("all_product_sales_data_stackplot.pdf", dpi=300,bbox_inches="tight")

plt.show()
```
