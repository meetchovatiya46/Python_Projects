```python
# movies_vs_tvshows_netflix

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("netflix_titles.csv")

# Remove rows with missing important values
df = df.dropna(subset=["type", "release_year", "rating", "country", "duration"])

# Count movies and TV shows
type_counts = df["type"].value_counts()

# Create bar chart
plt.figure(figsize=(6, 4))

plt.bar(
    list(type_counts.index),
    list(type_counts.values),
    color=["#E50914", "#221F1F"]
)

plt.title("Number of Movies VS T.V. Shows on Netflix")
plt.xlabel("Types")
plt.ylabel("Numbers")

plt.tight_layout()

# Save figures
plt.savefig("movies_vs_tvshows_netflix.pdf", dpi=300, bbox_inches="tight")
plt.savefig("movies_vs_tvshows_netflix.jpg", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# rating_count_netflix

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("netflix_titles.csv")

df=df.dropna(subset=["type", "release_year", "rating", "country", "duration"])

rating_count=df["rating"].value_counts()
plt.figure(figsize=(8,6))
plt.pie(rating_count, labels=rating_count.index, autopct="%1.1f%%", startangle=90)

plt.title("Number of Shows per Rating on Netflix", fontsize=14, fontweight="bold")


plt.savefig("rating_count_netflix.pdf", dpi=300, bbox_inches="tight")

plt.savefig("rating_count_netflix.jpg", dpi=300, bbox_inches="tight")

plt.show()
```

```python
# duration of movies in Netflix

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("netflix_titles.csv")

df=df.dropna(subset=["type", "release_year", "rating", "country", "duration"])

movie_duration=df[df["type"]=="Movie"].copy()
movie_duration["mins"]=movie_duration["duration"].str.replace("min","").astype(int)

plt.figure(figsize=(10,6))

plt.hist(movie_duration["mins"], bins=30, color="red", edgecolor="black")
plt.title("Duration of Movies in Netflix", fontsize=14, fontweight="bold")

plt.xlabel("Duration (mins)")
plt.ylabel("Number of Movies")

plt.savefig("duration_of_movies_in_netflix.pdf", dpi=300, bbox_inches="tight")

plt.savefig("duration_of_movies_in_netflix.jpg", dpi=300, bbox_inches="tight")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv("netflix_titles.csv")

# Drop missing values for required columns
df = df.dropna(subset=["type", "release_year", "rating", "country", "duration"])

# Get top 10 countries producing most shows/movies
countries = df["country"].value_counts().head(10)

# Create horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(countries.index, countries.values, color=plt.cm.plasma(range(10)))

# Add title and labels
plt.title("Top 10 Countries Producing Most Shows or Movies on Netflix", fontsize=14, fontweight='bold')
plt.xlabel("Number of Shows/Movies", fontsize=12)
plt.ylabel("Countries", fontsize=12)

# Add grid and adjust layout
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()

# Reverse order so the highest is on top
plt.gca().invert_yaxis()

plt.savefig("top_10_countries.pdf", dpi=300, bbox_inches="tight")
plt.savefig("top_10_countries.jpg", dpi=300, bbox_inches="tight")
# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read dataset
df = pd.read_csv("netflix_titles.csv")

# Drop missing values for necessary columns
df = df.dropna(subset=["type", "release_year"])

# Group by year and type, count how many titles each year
yearly_counts = df.groupby(["release_year", "type"]).size().unstack(fill_value=0)

# Sort by year (just in case)
yearly_counts = yearly_counts.sort_index()

# Create line chart
plt.figure(figsize=(10,6))
plt.plot(yearly_counts.index, yearly_counts["Movie"], marker='o', label='Movies', color="#FF6666", linewidth=2)
plt.plot(yearly_counts.index, yearly_counts["TV Show"], marker='s', label='TV Shows', color="#66B2FF", linewidth=2)

# Add labels, title, and legend
plt.title("Netflix: Movies vs TV Shows Over the Years", fontsize=14, fontweight='bold')
plt.xlabel("Release Year", fontsize=12)
plt.ylabel("Number of Titles", fontsize=12)
plt.legend(title="Type", fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.savefig("movies_vs_tvshows_over_years.pdf", dpi=300, bbox_inches="tight")
plt.savefig("movies_vs_tvshows_over_years.jpg", dpi=300, bbox_inches="tight")
# Show plot
plt.show()
```
